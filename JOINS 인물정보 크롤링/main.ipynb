````python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
````

````python
search_url = "http://people.joins.com/Search/"
````

````python
params = {'q': '문재인'}
````

````python
html = requests.get(search_url, params=params).text
html
````

````python
soup = BeautifulSoup(html, 'html.parser')
````

````python
for tag in soup.select('.result_list .img_listarea > li > dl'):
    a_tag = tag.find('a')
    url = urljoin(search_url, a_tag['href'])
    name = a_tag.text

    attrs = {'url': url, 'name': name}
    for dd_tag in tag.select('dd'):
        key = dd_tag.find('strong').text
        value = dd_tag.find('span').text
        attrs[key] = value

    print(attrs)
````

## 위 코드를 함수형태로 정리

````python
def search_people(q):
    search_url = "http://people.joins.com/Search/"    
    params = {'q': q}
    html = requests.get(search_url, params=params).text
    soup = BeautifulSoup(html, 'html.parser')
    
    result = []
    for tag in soup.select('.result_list .img_listarea > li > dl'):
        a_tag = tag.find('a')
        url = urljoin(search_url, a_tag['href'])
        name = a_tag.text

        attrs = {'url': url, 'name': name}
        for dd_tag in tag.select('dd'):
            key = dd_tag.find('strong').text
            value = dd_tag.find('span').text
            attrs[key] = value

        result.append(attrs)

    return result
````

````python
search_people('문재인')
````

````python
search_people('안철수')
````
