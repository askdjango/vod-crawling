주의 : 불법적인 크롤링에 대한 책임은 전적으로 본인에게 있습니다.

````python
import csv
import time
from itertools import count
import pandas
import requests
from bs4 import BeautifulSoup
````

````python
def get_page(page):
    url = "http://www.g2b.go.kr:8101/ep/tbid/tbidList.do"
    params = {
        "currentPageNo": page,

        "taskClCds": "",
        "bidNm": "",
        "searchDtType": 1,
        "fromBidDt": "2017/03/25",
        "toBidDt": "2017/04/24",
        "fromOpenBidDt": "",
        "toOpenBidDt": "",
        "radOrgan": 1,
        "instNm": "",
        "area": "",
        "regYn": "Y",
        "bidSearchType": 1,
        "searchType": 1,
    }

    html = requests.get(url, params=params).text
    soup = BeautifulSoup(html, 'html.parser')
    
    table_tag = soup.select('#container .results > table')[0]

    headers = [th_tag.text for th_tag in table_tag.select('thead > tr > th')]

    rows = []
    for tr_tag in table_tag.select('tbody > tr'):
        cols = [td_tag.text for td_tag in tr_tag.select('td')]
        if len(cols) > 1:
            rows.append(cols)
    
    return {
        'headers': headers,
        'rows': rows,
    }
````

````python
headers = None
rows = []
for page in count(1):
    print('try to page = {}'.format(page))
    data = get_page(page)
    if not data['rows']:
        print('페이지 끝에 도달')
        break
    if headers is None:
        headers = data['headers']
    rows.extend(data['rows'])
    time.sleep(0.5)
    
    if page >= 10:
        print('10페이지까지만 진행하겠습니다.')
        break
````

````python
with open('result.csv', 'wt', encoding='utf8') as f:
    writer = csv.writer(f)
    writer.writerow(headers)
    writer.writerows(rows)
````
